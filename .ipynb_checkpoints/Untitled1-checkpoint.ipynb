```python
import pandas as pd
```

```python
ratings = pd.read_csv("baseline-rs/datasets/ratings.csv", encoding='"ISO-8859-1"')
```

```python
ratings
```

```python
user_mappings = {}
movie_mappings = {}
updated_rating = ratings.copy()
def average_ratings():
    for index, row in ratings.iterrows():
        if row['movieId'] not in movie_mappings:
            movie_mappings[row['movieId']] = {}

        if row['userId'] not in user_mappings:
            user_mappings[row['userId']] = [1, row['rating']]
        else:
            user_mappings[row['userId']][0] += 1
            user_mappings[row['userId']][1] += row['rating']

    for index, row in ratings.iterrows():
        updated_ratings.at[index,'rating'] -= (user_mappings[row['userId']][1] / user_mappings[row['userId']][0])
        
```

```python
average_ratings()
```

```python
def create_item_similarity_matrix():
    for movie_id_1 in movie_mappings:
        for movie_id_2 in movie_mappings:
            if movie_id_2 == movie_id_1:
                continue
            for user_id in user_mappings:
                row_movie_1 = updated_ratings.loc(
                    updated_ratings['userId'] == user_id &&
                    updated_ratings['movieId'] == movie_id_1
                )
                row_movie_2 = updated_ratings.loc(
                    updated_ratings['userId'] == user_id &&
                    updated_ratings['movieId'] == movie_id_1
                )
                
            
```

```python
create_item_similarity_matrix()
```
